```python
neurons=[32,128]
penalties=[0.01, 0.1,1,10,100,'x']
for neurons_j in neurons:
    for penalty_i in penalties:
        for i in range(0 ,1):
            file_name='RUA_de5_tcn_clf9_n'+str(neurons_j)+'_pe'+str(penalty_i)+"_"+str(i)
            Target_DirPath = 'Deep-Learning-in-Python-master/'
            df = pd.read_csv(Target_DirPath + file_name+'.csv',index_col=0)
            #df['Date'] = pd.to_datetime(df["Date"])
            df_sum = df
            print("df_sum:\n\n")
            print(df_sum)
        for i in range(1 ,10):
            file_name='RUA_de5_tcn_clf9_n'+str(neurons_j)+'_pe'+str(penalty_i)+"_"+str(i)
            Target_DirPath = 'Deep-Learning-in-Python-master/'
            df = pd.read_csv(Target_DirPath + file_name+'.csv',index_col=0)
            #df['Date'] = pd.to_datetime(df["Date"])
            df_sum += df
            print("df_sum:\n\n")
            print(df_sum)
        df_sum = df_sum.div(10)
        print('real_df_sum\n\n\n')
        print(df_sum)
        ave_file_name='m_'+'RUA_de5_tcn_n'+str(neurons_j)+'_pe'+str(penalty_i)
        
        if (not (os.path.exists(Target_DirPath))):
                os.makedirs(Target_DirPath)
        df_sum.to_csv(Target_DirPath + ave_file_name+'.csv', mode='w', header=True, index=True)
```
